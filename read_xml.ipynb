## Descargando datos en XML

1. Descarga el archivo como xml y revisa el arbol que se genera 


```python
#import libraries
import requests
import xml.dom.minidom
import pandas as pd
import numpy as np
import xml.etree.cElementTree as et
from inspect import getmembers , isclass , isfunction
```

```python
#Save file from url

url = "https://raw.githubusercontent.com/mcd-unison/ing-caract/main/ejemplos/integracion/ejemplos/wikipedia-poetas.xml"

response = requests.get(url)
with open('poetas.xml', 'wb') as file:
    file.write(response.content)
```

```python
# read xml to root
tree = et.parse('./poetas.xml')
root = tree.getroot()
```

```python
#read author and text data

author1 = []
text1 = []

for autor in root.findall('{http://www.mediawiki.org/xml/export-0.10/}page'):
    
    # encontrar autor, y titulo de las obras
    title =  autor.find('{http://www.mediawiki.org/xml/export-0.10/}title')
    revision =  autor.find('{http://www.mediawiki.org/xml/export-0.10/}revision')
    texto = revision.find('{http://www.mediawiki.org/xml/export-0.10/}text')
    
    author1.append(title.text)
    text1.append(texto.text)
```

```python
#Definir DF con datos de autor, uniendolo con texto

daut = pd.DataFrame(author1)
daut.columns = ['Author']

dtexto = pd.DataFrame(text1)
dtexto.columns = ['Full_Text']

df_autores = pd.concat([daut, dtexto], axis=1, join='inner')
df_autores

```

```python
# Encontrar obras con Regex

import re
novela = []

for text in df_autores['Full_Text']:
    novela_found = re.findall("(==\s*Novela\w*\s*==)(.*?)(\n==)", text, flags = re.DOTALL)
    novela.append(novela_found)
novela
```

```python
df_novela = pd.DataFrame(novela)
df_novela.columns = ['Novela']
df_novela
```

```python
micro = []
for text in df_autores['Full_Text']:
    micro_found = re.findall("(==\\s*Microrrelatos\w*\s*==)(.*?)(\n==)", text, flags = re.DOTALL)
    micro.append(micro_found)
micro

```

```python
df_micro = pd.DataFrame(micro)
df_micro.columns = ['Microrrelatos']
df_micro
```

```python
#Unir 3 df, autores, novela, y microrrelatos
df_autores = pd.concat([df_autores, df_novela, df_micro], axis=1, join='inner')
df_autores
```

4. Guarda el DataFrame generado como un archivo de tipo parquet


```python
df_autores.to_parquet('poetas_df.parquet') 
```
