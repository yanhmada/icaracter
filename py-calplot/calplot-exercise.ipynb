```python
#Importing libraries
import pandas as pd
import calplot
import warnings
warnings.filterwarnings('ignore')
```

This data was originally obtained from https://www.gob.mx/conagua

```python
#loading data into Dataframe
df_tem = pd.read_csv('max_hmo.csv', delimiter=",")
df_tem.tail(2)
```

In order to create our heatmap we need to verify that date column is correctly formatted

```python
df_tem.info()
```

```python
df_tem.head(3)
```

```python

df_tem['Date'] =  pd.to_datetime(df_tem['Date'], format='%d/%m/%Y')
```

```python
df_tem.info()
```

```python
df_tem = df_tem.set_index('Date') #set datetime column as index
df_tem.head(3)
```

```python
calplot.calplot(df_tem['Max_temperatures'], cmap="coolwarm", colorbar=False);
```

We can add a colorbar by setting the value to true, and use the different pattern colors available for matplotlib.

```python
calplot.calplot(df_tem['Max_temperatures'], cmap="BuGn", colorbar=True);
```

We can also modify the size of figure, add suptitles, modify linecolors or add fillcolors.

```python

calplot.calplot(df_tem['Max_temperatures'], cmap= "BuGn", colorbar=True, 
             linecolor ="white" , fillcolor= "gray", figsize = (8, 4), suptitle = "Temperature by Month and Year");
```

For the complete options please visit https://calplot.readthedocs.io/en/latest/
